# **Data** **Preprocessing**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

# Sample data (representative columns)
data = pd.DataFrame({
    'PLID': ['PLID_4964', 'PLID_1123', 'PLID_1912', 'PLID_1746'],
    'age': [23, 18, 16, 27],
    'gender': ['M', 'M', 'F', 'M'],
    'country': ['Ukraine', 'Russia', 'Spain', 'Chile'],
    'height': ["6'2", "5'10", "6'2", "5'11"],
    'weight': ['176lbs', '148lbs', '172lbs', '165lbs'],
    'ball_contr': [77, 54, 34, 79],
    'body_refl': [58, 69, 43, 73],
    'body_bala': [47, 70, 45, 76],
    'jumping_spenalties_mental_st': [59, 62, 69, 86],
    'goalkeepi': [41, 53, 49, 81],
    'defending': [73, 43, 45, 80],
    'passing_s/dribbling_shot_accu': [13, 10, 10, 3],
    'body_stre': [69, 30, 57, 85],
    'max_runn': [70, 50, 37, 80],
    'stron': [36, 30, 18, 40],
    'selection': [1, 0, 0, 1]
})

# Converting height to numeric (inches for simplicity)
data['height'] = data['height'].apply(lambda x: int(x.split("'")[0]) * 12 + int(x.split("'")[1]) if "'" in x else None)

# Converting weight to numeric
data['weight'] = data['weight'].str.replace('lbs', '').astype(float)

# One-hot encoding for existing categorical columns
data = pd.get_dummies(data, columns=['gender', 'country'], drop_first=True)

# Scaling numerical features
scaler = StandardScaler()
numerical_features = ['age', 'height', 'weight', 'ball_contr', 'body_refl', 'body_bala', 'jumping_spenalties_mental_st', 'goalkeepi', 'defending', 'passing_s/dribbling_shot_accu', 'body_stre', 'max_runn', 'stron']
data[numerical_features] = scaler.fit_transform(data[numerical_features])

# Plotting the correlation heatmap
numeric_data = data.select_dtypes(include=[float, int])
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, train_test_split

# Assuming data is already preprocessed and available as 'data'
# Splitting the data
X = data.drop('selection', axis=1)
y = data['selection']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)  # Adjusted split to ensure enough samples for training

# Parameter grid for hyperparameter tuning
param_grid = {
    'C': [0.1, 1, 10, 100],
    'solver': ['lbfgs', 'liblinear']
}

# Performing grid search with reduced cv value
grid_search = GridSearchCV(LogisticRegression(), param_grid, cv=2)  # Reduced cv to 2
grid_search.fit(X_train, y_train)

# Best parameters
print("Best Parameters:", grid_search.best_params_)

# Evaluating the tuned model
best_model = grid_search.best_estimator_
y_pred_tuned = best_model.predict(X_test)

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("Tuned Accuracy:", accuracy_score(y_test, y_pred_tuned))
print("Tuned Classification Report:\n", classification_report(y_test, y_pred_tuned))
print("Tuned Confusion Matrix:\n", confusion_matrix(y_test, y_pred_tuned))
```

```python
import numpy as np

# Getting feature importance
importance = np.abs(best_model.coef_[0])
feature_names = X.columns
feature_importance = pd.DataFrame({'Feature': feature_names, 'Importance': importance}).sort_values(by='Importance', ascending=False)

print(feature_importance)
```
